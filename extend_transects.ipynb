```python
import geopandas as gpd
import math
from shapely.geometry import Point, LineString, MultiPoint, GeometryCollection
```

```python
transects = gpd.read_file("transects.geojson")
transects
```

```python
transects.to_crs(2193).length.describe()
```

```python
nz = transects[transects.site_id.str.startswith("nz")]
nz
```

```python
nz.to_crs(2193, inplace=True)
```

```python
nz.geometry.iloc[0].length
```

```python
def calculate_new_coordinates(old_x, old_y, bearing, distance):
    bearing_radians = math.radians(bearing)
    new_x = old_x + (distance * math.sin(bearing_radians))
    new_y = old_y + (distance * math.cos(bearing_radians))
    point = Point(new_x, new_y)
    assert not point.is_empty
    return point

def extend(row):
    p1, p2 = row.geometry.coords
    angle = row.orientation
    # Landward
    p1 = calculate_new_coordinates(p1[0], p1[1], angle, -200)
    # Seaward
    p2 = calculate_new_coordinates(p2[0], p2[1], angle, 400)
    return LineString([p1, p2])

nz.geometry = nz.apply(extend, axis=1)
```

```python
transects.geometry.update(nz.geometry.to_crs(transects.crs))
```

```python
transects.to_file("transects_extended.geojson")
```
